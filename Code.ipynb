```python
#5.1

#CHI-SQUARED TEST
#The aim of this chi-square test is to conclude whether the two variables(danceability and spotify streaming hours ) are related 
#to each other not. 
from scipy.stats import chi2_contingency
#Feature 1 : Danceability

feat1= [df["Danceability"], df["Spotify_stream"]]
stat, p, dof, expected = chi2_contingency(feat1)
 # interpret p-value
alpha = 1   #Since the values in col spotify_stream are more than the range 0-1, hence selecting  a higher value of aplha
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
    
#This shows that the no. of streaming hours of a track on Spotify is dependent on the music attribute - Danceability
```

## 6.1 ML technique 1 + Justification

```python
#ML TECHNIQUE 1 : CLUSTERING(k means)
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Scaling
#The clustering code starts with the normalization of the columns with a scaling function. As we observed, all features that 
#are provided by Spotify range between 0 and 1, except 2 of them: loudness, and tempo. We scale them to [0,1] in order to make
#them compatible with other columns in the vector space.

#X as defined above contains the numerical features

X['Loudness'] = ((X['Loudness'] - X['Loudness'].min()) / (X['Loudness'].max() - X['Loudness'].min()))
X['Tempo'] = ((X['Tempo'] - X['Tempo'].min()) / (X['Tempo'].max() - X['Tempo'].min()))

# Standardize the features
scaler = StandardScaler()
df_feat_scaled = scaler.fit_transform(X)
```

```python
#Here, we are doing PCA as the number of features for our dataset is quite high and  many of them are correlated. 
#PCA reduces the number of features and feature-vectors are re-engineered in such a way that it contains less number of features
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_feat_scaled)
```

```python
# Determining the cluster size
score_list = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(principalComponents)
    score_list.append(kmeans.inertia_)

plt.plot(range(1, 11), score_list)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Scores list')
plt.show()
                                             
```

```python
# Training and Predicting
#From elbow method, we take the optimal clusters as 4

kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
X['Popularity_cluster'] = kmeans.fit_predict(principalComponents)
```

```python
#Joining the features column, also containing the Popularity_cluster column with the target column Y
df_new = X.join(Y)
df_new.head()
```
